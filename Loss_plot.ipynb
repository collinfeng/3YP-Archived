```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plot losses
loss_cnn1 = np.load("Results/ELBO/VAE64CNN1—200_Result.npy")
loss_cnn2 = np.load("Results/ELBO/VAE64CNN2—200_Result.npy")
loss_vgg = np.load("Results/ELBO/VAE64VGG—200_Result.npy")
loss_res = np.load("Results/ELBO/VAE64Res—200_Result.npy")
```

```python
# normalisation
# ELBO train	ELBO val
# 20994.326868693035	26255.757582720587
# 40170.58241780599	44466.8671875
# 33628.26915486654	36067.70921415441
# 28532.20635986328	30669.745059742647
nvgt = loss_vgg[0, 199] / 20994.326868693035
nvgv = loss_vgg[2, 199] / 26255.757582720587
nret = loss_res[0, 199] / 40170.58241780599
nrev = loss_res[2, 199] / 44466.8671875
nc1t = loss_cnn1[0, 199] / 33628.26915486654
nc1v = loss_cnn1[2, 199] / 36067.70921415441
nc2t = loss_cnn2[0, 199] / 28532.20635986328
nc2v = loss_cnn2[2, 199] / 30669.74505974264
```

```python
plt.plot(loss_vgg[0, 1:100]/nvgt, label = "loss_vgg_t")
plt.plot(loss_vgg[2, 1:100]/nvgv, label = "loss_vgg_v")
plt.plot(loss_res[0, 1:100]/nret, label = "loss_res_t")
plt.plot(loss_res[2, 1:100]/nrev, label = "loss_res_v")
plt.plot(loss_cnn1[0, 1:100]/nc1t, label = "loss_cnn1_t")
plt.plot(loss_cnn1[2, 1:100]/nc1v, label = "loss_cnn1_v")
plt.plot(loss_cnn2[0, 1:100]/nc2t, label = "loss_cnn2_t")
plt.plot(loss_cnn2[2, 1:100]/nc2v, label = "loss_cnn2_v")
plt.legend(loc="upper right")
plt.savefig('figures/Prototype/elbo-loss.eps')
```
